```python
import pandas as pd
import os

folder_path = r"C:\Users\e410675\OneDrive - London South Bank University\LSBU - Jose's Collaborative Projects\Norfolk and Waveney Demand Modelling\New Data\Delayed Transfers of Care Data 2010-2020\CSV Files"
combined_data = []

# List of files that should use row 1 as header (i.e., skiprows=0)
header_row_1_files = [
    'Delayed Transfers of Care 2018-04.csv',
    'Delayed Transfers of Care 2018-05.csv',
    'Delayed Transfers of Care 2018-06.csv',
    'Delayed Transfers of Care 2018-07.csv',
    'Delayed Transfers of Care 2018-08.csv',
    'Delayed Transfers of Care 2018-09.csv',
    'Delayed Transfers of Care 2018-10.csv',
    'Delayed Transfers of Care 2018-11.csv',
    'Delayed Transfers of Care 2018-12.csv',
    'Delayed Transfers of Care 2019-01.csv',
    'Delayed Transfers of Care 2019-02.csv',
    'Delayed Transfers of Care 2019-03.csv',
    'Delayed Transfers of Care 2019-04.csv',
    'Delayed Transfers of Care 2019-05.csv',
    'Delayed Transfers of Care 2019-06.csv',
    'Delayed Transfers of Care 2019-07.csv',
    'Delayed Transfers of Care 2019-08.csv',
    'Delayed Transfers of Care 2019-09.csv',
    'Delayed Transfers of Care 2019-10.csv',
    'Delayed Transfers of Care 2019-11.csv',
    'Delayed Transfers of Care 2019-12.csv',
    'Delayed Transfers of Care 2020-01.csv',
    'Delayed Transfers of Care 2020-02.csv'
]

for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)

        # Determine which row to use as header
        if file_name in header_row_1_files:
            data = pd.read_csv(file_path, skiprows=0)
        else:
            data = pd.read_csv(file_path, skiprows=4)  # Skip first 4 rows to use 5th row as header

        data['source file name'] = file_name  # Add source file column
        combined_data.append(data)

# Concatenate all dataframes
combined_data = pd.concat(combined_data, axis=0, ignore_index=True)

```

# Clean the dataset 

### Add a column  with month and yearf

```python


import pandas as pd


combined_data['Year final'] = combined_data['source file name'].str.extract(r'(\d{4})')
combined_data['Month'] = combined_data['source file name'].str.extract(r'-(\d{2})')

combined_data['Year-Month'] = combined_data['Year final'] + '-' + combined_data['Month']
```

### Combine all the columns that have the sames infromation but the columns name change over the time

```python
import pandas as pd


# Fill unified column by checking the first non-null (or non-empty) value
combined_data["Provider Parent Org Code final"] = combined_data["Provider SHA Code"].combine_first(combined_data["Provider Parent Org Code"])

# Drop the original columns that were replaced
combined_data.drop(columns=["Provider SHA Code", "Provider Parent Org Code"], inplace=True)



# Fill unified column by checking the first non-null (or non-empty) value
combined_data["Provider Parent Name final"] = combined_data["Provider SHA Name"].combine_first(combined_data["Provider Parent Name"])
combined_data.drop(columns=["Provider SHA Name", "Provider Parent Name"], inplace=True)


combined_data["Period Name final"] = combined_data["Period Name"].combine_first(combined_data["Period"])
combined_data.drop(columns=["Period Name", "Period"], inplace=True)

```

```python
combined_data["NHS DTOC Beds final"] = combined_data["NHS A SUM"].combine_first(combined_data["NHS DTOC Beds"]).combine_first(combined_data["NHS DTOC beds"]).combine_first(combined_data["NHS DToC beds"])
combined_data.drop(columns=["NHS A SUM", "NHS DTOC Beds", "NHS DTOC beds", "NHS DToC beds"], inplace=True)

combined_data["Social Care DTOC Beds final"] = combined_data["Social Care A SUM"].combine_first(combined_data["Social Care DTOC Beds"]).combine_first(combined_data["Social Care DTOC beds"]).combine_first(combined_data["Social Care DToC beds"])
combined_data.drop(columns=["Social Care A SUM", "Social Care DTOC Beds", "Social Care DTOC beds", "Social Care DToC beds"], inplace=True)

combined_data["Both DTOC Beds final"] = combined_data["Both A SUM"].combine_first(combined_data["Both DTOC Beds"]).combine_first(combined_data["Both  DTOC beds"]).combine_first(combined_data["Both DToC beds"]).combine_first(combined_data["Both DTOC beds"])
combined_data.drop(columns=["Both A SUM", "Both DTOC Beds", "Both  DTOC beds", "Both DToC beds", "Both DTOC beds"], inplace=True)
```

```python
combined_data["NHS final"] = combined_data["NHS B SUM"].combine_first(combined_data["NHS"])
combined_data.drop(columns=["NHS B SUM", "NHS"], inplace=True)

combined_data["Social Care final"] = combined_data["Social Care B SUM"].combine_first(combined_data["Social Care"])
combined_data.drop(columns=["Social Care B SUM", "Social Care"], inplace=True)

combined_data["Both B final"] = combined_data["Both B SUM"].combine_first(combined_data["Both B"])
combined_data.drop(columns=["Both B SUM", "Both B"], inplace=True)
```

### Organize columns names

```python
# Desired column order
new_column_order = [
    "Year",
    "Period Name final",
    "Year final",
    "Month",
    "Year-Month",
    "Provider Parent Org Code final",
    "Provider Parent Name final",
    "Provider Org Code",
    "Provider Org Name",
    "Local Authority Code",
    "Local Authority Name",
    "Acute Or Non Acute Description",
    "Reason For Delay",
    "NHS DTOC Beds final",
    "NHS final",
    "Social Care DTOC Beds final",
    "Social Care final",
    "Both DTOC Beds final",
    "Both B final",
    "source file name"
]

# Reorder the columns
combined_data = combined_data[new_column_order]

```

### Drop some columns

```python
combined_data = combined_data.drop(columns=["Year", "Period Name final"])
```
